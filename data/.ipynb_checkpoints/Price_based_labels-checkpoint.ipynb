```python
stock_close['move_avg_1d'] = ((stock_close.ibm_1d/stock_close.ibm)
                              /(stock_close.dji_1d/stock_close.dji))

stock_close['move_avg_2d'] = ((stock_close.ibm_2d/stock_close.ibm_1d)
                              /(stock_close.dji_2d/stock_close.dji_1d))

stock_close['move_avg_3d'] = ((stock_close.ibm_3d/stock_close.ibm_2d)
                              /(stock_close.dji_3d/stock_close.dji_2d))

stock_close['move_avg'] = (stock_close.move_avg_1d)
```

```python
stock_close.head(22)
```

```python
plt.hist(stock_close.move_avg, bins=50);
```

```python
third1 = np.percentile(np.array(stock_close.move_avg.iloc[:-3]), 33)
```

```python
third2 = np.percentile(np.array(stock_close.move_avg.iloc[:-3]), 66)
```

```python
def price_based_labeling(average_index):
    if average_index < third1:
        return 'bad'
    elif average_index > third2:
        return 'good'
    else:
        return 'neutral'

stock_close['price_based_label'] = stock_close.move_avg.apply(price_based_labeling)
```

```python
stock_close.loc[(stock_close.ibm_1d>stock_close.ibm)
                &(stock_close.ibm_2d>stock_close.ibm_1d),'price_based_label2'] = 'good'

stock_close.loc[(stock_close.ibm_1d<stock_close.ibm)
                &(stock_close.ibm_2d<stock_close.ibm_1d),'price_based_label2'] = 'bad'

stock_close.loc[(stock_close.price_based_label2!='good')
                &(stock_close.price_based_label2!='bad'),'price_based_label2'] = 'neutral'
```

### Creating sentences, finding relevant sentences

#### Replace lineend character with whitespace

```python
ibm_articles.text = ibm_articles.text.apply(lambda x: x.replace('\n', ' '))
ibm_articles.subhead = ibm_articles.subhead.apply(lambda x: x.replace('\n', ' ') if x!=None else x)
```

#### Find relevant sentences and store in new column

```python
def find_relevant_sentence(text, keywords):
    sentences = []
    for sentence in find_sentences(text):
        for keyword in keywords:
            if keyword in sentence:
                sentences.append(sentence)
                break
    return sentences

def find_ibm(text):
    return find_relevant_sentence(text, ['IBM', 'International Business Machine'])
```

```python
ibm_articles['sentences'] = ibm_articles.text.apply(find_ibm)
```

#### Creating relevant text: keeping just the relevant sentences if the company is not mentioned in the head

```python
ibm_articles.loc[(ibm_articles.relevant==1)|(ibm_articles.sub_relevant==1),'relevant_text'] = ibm_articles.text
ibm_articles.loc[(ibm_articles.relevant==0)&(ibm_articles.sub_relevant==0),'relevant_text'] = ibm_articles.sentences
```
